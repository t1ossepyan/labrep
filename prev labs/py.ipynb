# Python Syntax Basics

## Python Comments
Comments in Python start with a hash (#) symbol.

```python
# This is a comment
print('Hello, World!')  # This is also a comment
```

## Python Variables
Variables in Python are declared with an assignment operator (=).

```python
x = 5
y = 'Hello'
```

## Variable Names
Variable names in Python can contain letters, digits, and underscores. They cannot start with a digit.

```python
my_variable = 123
another_variable = 'example'
_internal_var = True
# 2nd_variable = 'invalid'  # Invalid variable name
```

## Assign Multiple Values
Python allows you to assign multiple variables in a single line.

```python
a, b, c = 1, 2, 3
```

## Python Strings
Strings in Python are enclosed in either single quotes ('') or double quotes ("").

```python
message = 'Hello, World!'
name = "Alice"
```

## Output Variables
You can output variables using print statements.

```python
x = 10
print('The value of x is:', x)
```

## Global Variables
Global variables are declared outside of any function and can be accessed globally.

```python
global_var = 'global'
def print_global():
    print('Global variable value:', global_var)
```

## Python Data Types
Python supports several built-in data types, including numbers, strings, lists, tuples, dictionaries, etc.

## Python Numbers
Python supports integers, floats, and complex numbers.

```python
int_num = 10
float_num = 3.14
complex_num = 1 + 2j
```

## Python Casting
Casting in Python is done using constructor functions like int(), float(), str(), etc.

```python
x = int(3.14)  # x will be 3
y = float('3.14')  # y will be 3.14
z = str(123)  # z will be '123'
```
