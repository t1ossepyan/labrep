# Основы синтаксиса Python

## 1. Boolean (Булевы значения)
Булевы значения могут быть `True` или `False`.
```python
x = True
y = False
print(x, y)  # Выведет: True False
```

## 2. Операторы (Operators)
Python поддерживает арифметические, логические, побитовые и другие операторы.
```python
a = 10
b = 3
print(a + b)  # Сложение
print(a - b)  # Вычитание
print(a * b)  # Умножение
print(a / b)  # Деление
print(a % b)  # Остаток от деления
print(a ** b) # Возведение в степень
print(a // b) # Целочисленное деление
```

## 3. Списки (Lists)
Списки могут содержать элементы разных типов.
```python
my_list = [1, 2, 3, 'hello', True]
print(my_list[0]) # Доступ по индексу
my_list.append(4) # Добавление элемента
print(my_list)
```

## 4. Кортежи (Tuples)
Кортежи похожи на списки, но неизменяемые.
```python
my_tuple = (10, 20, 30)
print(my_tuple[1]) # Доступ по индексу
```

## 5. Множества (Sets)
Множества хранят уникальные значения.
```python
my_set = {1, 2, 3, 4, 4, 4}
print(my_set)  # {1, 2, 3, 4}
```

## 6. Словари (Dictionaries)
Словари хранят пары ключ-значение.
```python
my_dict = {'name': 'Alice', 'age': 25}
print(my_dict['name']) # Доступ по ключу
my_dict['age'] = 26 # Изменение значения
print(my_dict)
```

## 7. Условные операторы (If...Else)
```python
x = 10
if x > 5:
    print('x больше 5')
elif x == 5:
    print('x равно 5')
else:
    print('x меньше 5')
```

## 8. Цикл While
```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

## 9. Цикл For
```python
for num in range(5):
    print(num)
```

## 10. Функции (Python Functions)
```python
def greet(name):
    return f'Hello, {name}!'
print(greet('Alice'))
```

## 11. Лямбда-функции (Python Lambda)
```python
square = lambda x: x * x
print(square(5))
```

## 12. Массивы (Python Arrays)
```python
import array
arr = array.array('i', [1, 2, 3, 4])
print(arr[2])
```

## 13. Классы и Объекты (Python Classes/Objects)
```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
p = Person('Alice', 30)
print(p.name, p.age)
```

## 14. Наследование (Python Inheritance)
```python
class Employee(Person):
    def __init__(self, name, age, position):
        super().__init__(name, age)
        self.position = position
e = Employee('Bob', 40, 'Manager')
print(e.name, e.position)
```
